## Transform

### Удаение пустых столбцов

```python
df = df.drop(df.columns[5:], axis=1)
```

### Столбец - код

```python
df[['Code', 'Data']] = df['Код'].str.split(',', n=1, expand=True)
df.head()
```

### Столбец - цвет

```python
def extract_color(text):
  text = str(text).strip()
  # if color in ()
  result = re.search(r"\((.*?)\)\s*$", text)
  if result:
    return result.group(1).strip()

  #if color after / or //
  result = re.search(r"/+\s*([^/]+)$", text)
  if result:
    return result.group(1).strip()

def remove_text(row, column_name):
    text = str(row['Data']).strip()
    value = row[column_name]

    if value:
        text = text.replace(value, '')
        text = ' '.join(text.split())
        text = text.replace('()', '').replace('( )', '')
        text = text.rstrip('/').strip()

    return text.strip()

df['Color'] = df['Data'].apply(extract_color)
df['Data'] = df.apply(remove_text, axis=1, column_name='Color')
```

### Столбец - размер

```python
def extract_size(text):
  text = str(text).strip()
  size_patterns = [
      r'(\d+\s*)(?:мм|см|м)?(\s*[*хx/]\s*)(\d+\s*)(?:мм|см|м)?(\s*[*хx/]\s*)?(\d+\s*)?(?:мм|см|м)?',
      r"(\d+\s*)(?:мм|см|м)",                           # 120 мм
      r"[A-Za-zА-Яа-я]\d+",                             # D120, J18
      r"(\d+\s*)(?:мм|см|м)(\s*[*хx/]\s*)(\d+\s*[˚°])"  # 80 мм * 45˚
      ]
  for p in size_patterns:
    result = re.search(p + r"\s*$", text, flags=re.IGNORECASE)
    if result:
      return result.group(0)
  return None

df['Size'] = df['Data'].apply(extract_size)
df['Data'] = df.apply(remove_text, axis=1, column_name='Size')
```

### Столбец - бренд

```python
def extract_brand(text):
  text = str(text).strip()
  brand_patterns = r"\b(Döcke|Docke|Dӧcke)\b"

  result = re.search(brand_patterns, text, flags=re.IGNORECASE)
  if result:
    return result.group(0)
  return None

df['Brand'] = df['Data'].apply(extract_brand)
df['Data'] = df.apply(remove_text, axis=1, column_name='Brand')
```

### Столбец - наименование

```python
def extract_name(text):
    text = str(text).strip()

    result = re.search(r'([а-яА-ЯёЁ][а-яА-ЯёЁ\s-]*[а-яА-ЯёЁ])', text)
    if result:
      return result.group(0)

df['Name'] = df['Data'].apply(extract_name)
```

Очистка столбца наименование от серий

```python
def remove_series(text):
  text=str(text).strip()

  series_list = ['PREMIUM', 'LUX', 'STANDARD','Standart', 'PIE', 'STEIN', 'KLINKER', 'FIX',
                 'GLASS', 'STAL', 'Comfort', 'EL', 'PLUS', 'BRICK', 'BERGART', 'BURG','FELS',
                 'DRAGON', 'STERN', 'COTA', 'EUROPA', 'KARAT', 'ROCKY', 'EURASIA', 'RA', 'DACHA',
                 'MONTERREY', 'NEXT', 'ROOT', 'BERG', 'MATRIX', 'SLATE', 'СОТА',
                 'ЭДЕЛЬ', 'НИЦЦА', 'САППОРО', 'ЖЕНЕВА', 'ГЕНУЯ', 'ЦЮРИХ', 'КЁЛЬН', 'ТЕТРИС','ФЛЕМИШ',
                 'СЛАНЕЦ', 'СЛАНЕЦ', 'ДЮФУР', 'АЛТАЙ', 'ШЕФФИЛД', 'КРОНА', 'ЛАБИРИНТ', 'ПЛИТКА', 'ШЕСТИГРАННИК',
                 'Монреаль']
  for s in series_list:
    text = re.sub(rf'\b{re.escape(s)}\b', '', text, flags=re.IGNORECASE)

  text = re.sub(r'^[\s/,]+', '', text)
  text = re.sub(r'[\s/,]+$', '', text)
  text = re.sub(r'\s+', ' ', text)

  return text.strip()

df['Name'] = df['Data'].apply(remove_series)
```

### Столбец - штук

```python
def extract_quantity(text):
  text = str(text).strip()
  result = re.search(r'\d+', text)
  if result:
    return int(result.group(0))

df['quantity'] = df['шт в упак'].apply(extract_quantity)
df['quantity'] = df['quantity'].fillna(0)
```

### Столбец - м3 на урпаковку

```python
# replace , to .
df['м3 упак'] = df['м3 упак'].astype(str).str.replace(',', '.')

df['m3_per_one'] = pd.to_numeric(df['м3 упак'], errors='coerce')
df['m3_per_one'] = df['m3_per_one'].fillna(0) #fill nan

df['m3_per_one'] = np.where(df['quantity'] == 0, 0, df['m3_per_one'] / df['quantity'])
```

### Столбец - брутто упаковки

```python
# replace , to .
df['брутто упак'] = df['брутто упак'].astype(str).str.replace(',', '.')

df['gross_per_one'] = pd.to_numeric(df['брутто упак'], errors='coerce')
df['gross_per_one'] = df['gross_per_one'].fillna(0) #fill nan

df['gross_per_one'] = np.where(df['quantity'] == 0, 0, df['gross_per_one'] / df['quantity'])
```
